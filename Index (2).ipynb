```python
#3
def no_of_digits(n):
    c = 0
    while n>0:
        c = c+1;
        n = n//10
    return c
n = int (input())
print(no_of_digits(n))
```

```python
#4
def factorial(n):
    return 1 if n<2 else n*factorial(n-1)
def mul(n):
    return 1 if n<2 else factorial(n)*mul(n-1)
n = int (input())
print(mul(n))
```
